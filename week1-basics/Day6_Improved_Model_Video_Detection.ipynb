```python
import os
from ultralytics import YOLO

print("📍 Current Directory:", os.getcwd())

# Search for best.pt
print("\n🔍 Searching for best.pt...")
found = False
for root, dirs, files in os.walk("."):
    if "best.pt" in files:
        full_path = os.path.join(root, "best.pt")
        print("✅ FOUND:", full_path)
        found = True

        # Try loading with this path
        try:
            model = YOLO(full_path)
            print("✅ Model loaded successfully using found path!")
            break
        except Exception as e:
            print("Failed to load with this path:", e)

if not found:
    print("❌ No best.pt file found in the current directory or subfolders.")
```

```python
from ultralytics import YOLO
import os

# ====================== LOAD MODEL ======================
model_path = "models/best.pt"

if os.path.exists(model_path):
    model = YOLO(model_path)
    print("✅ Model loaded successfully!")
else:
    print("❌ Model not found!")
    print("Current directory:", os.getcwd())
# ======================================================

print("\n" + "="*50)

# Skip validation if dataset is missing (optional)
dataset_path = "Waste-Detection-1/data.yaml"

if os.path.exists(dataset_path):
    print("Running Validation...")
    results = model.val(data=dataset_path)
else:
    print("⚠️ Dataset not found for validation (normal if running locally).")
    print("Skipping validation and moving to testing.")

print("\n✅ Day 6 Setup Complete!")
```

```python
# Test on an image
url = "https://images.unsplash.com/photo-1621451537084-482c73073a0f"
img_path = "test_waste.jpg"

import requests
response = requests.get(url, headers={"User-Agent": "Mozilla/5.0"})
with open(img_path, "wb") as f:
    f.write(response.content)

results = model.predict(img_path, conf=0.25)
results[0].show()
```

```python
# Video Detection Function
def detect_in_video(video_path, output_path="output_video.mp4"):
    results = model.predict(
        source=video_path,
        conf=0.25,
        save=True,
        name="video_detection"
    )
    print("✅ Video processing completed!")
    print(f"Output saved as: runs/detect/video_detection/{output_path}")
```

```python
# Export to different formats
model.export(format="onnx")      # Good for web/deployment
model.export(format="tflite")    # For mobile
```
